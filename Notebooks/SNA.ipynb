```python
import plotly.graph_objects as go
import numpy as np
import networkx as nx

# Create graph
G = nx.MultiDiGraph()

# Add edges with attributes
for _, row in CC_edge_table.iterrows():
    G.add_edge(row['Source'], row['Target'], weight=row['Communication_Score'], communication=row['Communication_Score'])

# Adding Node Attributes
Trust_dict = dict(zip(CT_edge_table['Target'], CT_edge_table['Trust_Score']))
OVR_dict = dict(zip(OVR_df['Stakeholder'], OVR_df['Overall Rating']))
satisfaction_dict = dict(zip(CS_edge_table['Target'], CS_edge_table['Satisfaction_Score']))

nx.set_node_attributes(G, satisfaction_dict, 'satisfaction')
nx.set_node_attributes(G, OVR_dict, 'overall_rating')
nx.set_node_attributes(G, Trust_dict, 'trust')

# setting Node trace attributes
# calculating Node position and distance based on Trust
pos = {'Consumers': (0, 0)}  # Consumers at the center
for _, row in CT_edge_table.iterrows():
    angle = 2 * np.pi * CT_edge_table.index.get_loc(_) / len(CT_edge_table)
    distance = 1 - row['Trust_Score']  # Invert so higher trust is closer
    pos[row['Target']] = (distance * np.cos(angle), distance * np.sin(angle))

node_x = [pos[0] for pos in pos.values()]
node_y = [pos[1] for pos in pos.values()]

#setting node size
node_size = [150 if Stakeholder == 'Consumers' else 150 * G.nodes[Stakeholder]['overall_rating'] for Stakeholder in G.nodes()]

#setting node color
node_color =  [G.nodes[Stakeholder]['satisfaction'] if Stakeholder != 'Consumers' else 0.5 for Stakeholder in G.nodes()]

# setting Node text
node_text = []
for Stakeholder in G.nodes():
    if Stakeholder != 'Consumers':
        # Get satisfaction and overall rating, defaulting to 'N/A' if they are missing
        Trust = G.nodes[Stakeholder].get('trust', 'N/A')
        satisfaction = G.nodes[Stakeholder].get('satisfaction', 'N/A')
        overall_rating = G.nodes[Stakeholder].get('overall_rating', 'N/A')

        # Ensure both satisfaction and overall_rating are formatted properly if numeric
        satisfaction_text = f"{satisfaction:.3f}" if isinstance(satisfaction, (int, float)) else satisfaction
        overall_rating_text = f"{overall_rating:.3f}" if isinstance(overall_rating, (int, float)) else overall_rating
        Trust_text = f"{Trust:.3f}" if isinstance(Trust, (int, float)) else Trust

        # Append the text for each stakeholder
        node_text.append(f"{Stakeholder}<br>Trust: {Trust_text}<br>Satisfaction: {satisfaction_text}<br>Overall Rating: {overall_rating_text}")
    else:
        node_text.append("Consumers")

# Create the node trace with satisfaction-based coloring
node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        colorscale='YlOrRd',  # Use a yellow-orange-red colorscale for satisfaction
        color=node_color,
        size=node_size,
        colorbar=dict(title='Satisfaction Score'),
        line=dict(width=2, color='DarkSlateGrey')
    ),
    text=node_text
)
# Prepare edges with varying widths based on communication scores and assign colors based on source node
edge_traces = []
arrows = []

# Example color map for source nodes (you can define your own)
color_map = {
    'Government': 'rgb(255, 0, 0)',  # Red
    'Utilities': 'rgb(0, 255, 0)',   # Green
    'Regulators': 'rgb(0, 0, 255)',  # Blue
    'Consumers': 'rgb(255, 255, 0)'  # Yellow
}

for edge in G.edges(data=True):
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]

    comm_score = edge[2]['communication']

    # Calculate the angle of the edge
    angle = np.arctan2(y1 - y0, x1 - x0)

    # Add a slight curve to separate bidirectional edges
    mid_x = (x0 + x1) / 2 + 0.05 * np.sin(angle)
    mid_y = (y0 + y1) / 2 - 0.05 * np.cos(angle)

    # Determine edge color based on the source node
    edge_color = color_map.get(edge[0], 'rgb(200,200,200)')  # Default to light gray if the source is not in the color_map

    edge_trace = go.Scatter(
        x=[x0, mid_x, x1],
        y=[y0, mid_y, y1],
        mode='lines',
        line=dict(width=1 + 5 * comm_score, color=edge_color),  # Use the source-based edge color
        hoverinfo='text',
        text=f"{edge[0]} to {edge[1]}<br>Communication Score: {comm_score:.3f}"
    )
    edge_traces.append(edge_trace)

    # Add arrow
    arrow = go.layout.Annotation(
        x=x1, y=y1,
        ax=mid_x, ay=mid_y,
        xref='x', yref='y', axref='x', ayref='y',
        showarrow=True,
        arrowhead=2,
        arrowsize=1.5,
        arrowwidth=1 + 5 * comm_score,
        arrowcolor=edge_color  # Use the source-based edge color for arrows
    )
    arrows.append(arrow)

# Create the figure
fig = go.Figure(data=[*edge_traces, node_trace],
                layout=go.Layout(
                    title='Renewable Energy Sector Network',
                    showlegend=False,
                    hovermode='closest',
                    margin=dict(b=20, l=5, r=5, t=40),
                    annotations=arrows,
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)
                ))

# Add annotations for node labels
for node, (x, y) in pos.items():
    fig.add_annotation(
        x=x, y=y,
        text=node,
        showarrow=False,
        font=dict(size=14)
    )

# Show the plot
fig.show()
fig.write_json("Renewable_Energy_SNA")
```
